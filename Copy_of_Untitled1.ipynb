```python
def my_function(x) :
  if x < 0 :
      return "Negative integer"
  else :
    factorial = 1
    for i in range (1 , x+1) :
      factorial = factorial * i
    return factorial
u = 5
my_function(u)
```

```python
def geometric_series (x,n) :
  sum = 0
  for i in range (n+1) :
     sum=sum + x**i
  return sum
```
